```python
import pandas as pd
```

```python
# Reading title.akas.tsv.gz
dff_titles = pd.read_csv(r"C:\Users\saipriya\Downloads\title.akas.tsv.gz",compression='gzip',sep='\t')
dff_titles.head()
```

```python
# Collecting only English Movies
dff_titles_eng = dff_titles[dff_titles['language']=='en']
```

```python
# Reading title.basics.tsv.gz
dff_basics_titles = pd.read_csv(r"C:\Users\saipriya\Downloads\title.basics.tsv.gz",compression='gzip',sep='\t')
dff_basics_titles.head()
```

```python
# Filtering only movies out of all types(short,movies...etc)
dff_basics_titles_movie = dff_basics_titles[dff_basics_titles['titleType'] == 'movie']
```

```python
#  Merging 
dff_actual = dff_basics_titles_movie.merge(dff_titles_eng, left_on='tconst', right_on='titleId')
dff_actual.shape
```

```python
dff_actual.head()
```

```python
# Creating anew DataFrame with only required
dff_actual_titles = dff_actual[['tconst' , 'title']]
```

```python
dff_actual_titles.head()
```

```python
# Reading title.ratings.tsv.gz 
dff_ratings_titles = pd.read_csv(r"C:\Users\saipriya\Downloads\title.ratings.tsv.gz",compression='gzip',sep='\t',usecols=('tconst' , 'averageRating'))
dff_ratings_titles.head()
```

```python
# Merging Titles with Ratings
dff_titles_ratings = dff_actual_titles.merge(dff_ratings_titles, on='tconst')
dff_titles_ratings.shape
```

```python
dff_titles_ratings.head()
```

```python
# name.basics.tsv.gz
dff_names = pd.read_csv(r"C:\Users\saipriya\Downloads\name.basics.tsv.gz",compression='gzip',sep='\t',usecols=('nconst' ,'primaryName' ))
dff_names.head()
```

```python
# title.principals.tsv.gz
dff_princ_titles = pd.read_csv(r"C:\Users\saipriya\Downloads\title.principals.tsv.gz",compression='gzip',sep='\t')
dff_princ_titles.head()
```

```python
# Only considering Cast(actor and actress)
dff_cast = dff_princ_titles[dff_princ_titles.category.isin(['actor' ,'actress'])]
dff_cast.head()
```

```python
# Merging names with cast
dff_names_cast = dff_cast.merge(dff_names,on='nconst')
dff_names_cast.head()
```

```python
# Merging titles with cast
dff_titles_cast = dff_actual_titles.merge(dff_names_cast,on='tconst')
dff_titles_cast.head()
```

```python
df = dff_titles_cast[['tconst','title' , 'category','primaryName']]
df.head()
```

```python
# combining the cast
df_title_cast = df.groupby(['tconst','title'])['primaryName'].apply(', '.join).reset_index()
df_title_cast.head()
```

```python
df_title_cast.drop('title' , axis=1, inplace=True)
```

```python
# Merging titles_cast with ratings
df_title_cast_ratings = df_title_cast.merge(dff_titles_ratings , on='tconst')
df_title_cast_ratings.head()
```

```python
df_title_cast_ratings.rename(columns={'primaryName':'cast','averageRating':'rating','title':'title_Name'}, inplace=True)
```

```python
movie_detail=df_title_cast_ratings.head(1000)
```

```python
movie_detail.to_csv(r"C:\Users\saipriya\Downloads\movies.tsv", index=False)
```
